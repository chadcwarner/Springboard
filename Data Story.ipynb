## Usage Percentage

Usage percentage is an estimate of the percentage of team plays used by a player while he was on the floor. In layman’s terms, it is the number of plays used by a specific player. What does “used” mean in that statement? It means the percentage of plays that ended with the player either shooting the ball, turning the ball over or getting to the free-throw line. It’s a measure of the percentage of the plays a player is involved in the offense which is very important.

```python
fig = plt.figure(figsize=(16, 6))

ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax.tick_params(labelrotation=90)
ax2.tick_params(labelrotation=90)

ax.title.set_text('LEBRON - Usage Percentage')
ax2.title.set_text('KOBE BRYANT - Usage Percentage')
ax.set(xlabel="Season",ylabel="Usage_Percentage")
ax2.set(xlabel="Season",ylabel="Usage_Percentage")
ax.plot(lebron_pts['Season'], lebron_pts['Usage_Percentage'])
ax2.plot(kobe_pts['Season'], kobe_pts['Usage_Percentage'])
plt.show()
```

This is interesting in the sense that it shows the Usage Percentage of both superstars over the entirety of their careers.  Lebron James has a higher floor but a lower ceiling than Kobe Bryant.  Neither side shows consistency but it appears that Lebron James had a higher Usage Percentage for a longer time than did Kobe Bryant, also a lot earlier in his career, he was more valuable to his team than was Kobe Bryant
