```python
#Determine "adopted" labels
adopted_or_not = []
for i in range(1,12001):
    logins = usr_engage.loc[usr_engage.user_id == i]
    resampled_data = logins.rolling('7d').count()
    if resampled_data.loc[resampled_data.visited >= 3].shape[0] > 0:
        adopt = 1
    else:
        adopt = 0
        
    adopted_or_not.append(adopt)
```

```python
usr['Adopted'] = adopted_or_not
usr.head()
```
